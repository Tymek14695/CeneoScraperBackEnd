```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
opinions = [filename.split(".")[0] for filename in os.listdir('opinions')]
for opinion in opinions:
    pass
product_code = opinions[0]
```

```python
opinions = pd.read_json(f'opinions/{product_code}.json')
opinions.stars = opinions.stars.apply(lambda s: s.split('/')[0].replace(',', '.')).astype(float)
```

```python
opinions_count = opinions.shape[0]
pros_count = opinions.pros.astype(bool).sum()
cons_count = opinions.cons.astype(bool).sum()
average_stars = opinions.stars.mean()
```

```python
stars_distr = opinions.stars.value_counts().reindex(np.arange(0, 5.5, 0.5), fill_value=0)
stars_distr.plot.bar()
plt.title(f"Histogram częstotliwości ocen produktu {product_code}")
plt.xlabel("liczba gwiazdek")
plt.ylabel("liczba opinii")
```

```python
pros_distr = opinions.pros.value_counts()
opinions.plot.scatter(pros_count, cons_count, s=2000, color='blue', alpha=0.5)
plt.title(f"Histogram częstotliwości ocen produktu {product_code}")
plt.xlabel("liczba gwiazdek")
plt.ylabel("liczba opinii")
```
